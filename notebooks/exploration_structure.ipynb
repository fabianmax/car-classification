```python
import os
import re
import pandas as pd
```

```python
def open_file_structure(path, clean=True):
    """
    Open and prepare file structure
    """
    with open(path) as f:
        structure = f.read()
        
    structure = [x.replace("'", "") for x in structure.split("', ")]
    
    if clean:
        # Tolower, remove special characters and whitespaces
        structure = [x.lower() for x in structure]
        structure = [re.sub('[^a-zA-Z0-9 \n\.]', '', x) for x in structure]
        structure = [x.replace(' ', '_') for x in structure]
        structure = [x.replace('__', '_') for x in structure]

    return structure
```

```python
def expand_column(df, column, col_names):
    """
    Extract columns for single string
    """
    df_expanded = pd.DataFrame(df[column].str.split('_', expand=True))
    df_expanded.columns = [*col_names, 'id']
    return df_expanded
```

```python
# Get jpg-files in raw folder
files = [file for file in os.listdir('raw') if file.endswith(".jpg")]

# Get file structure
file_name_components = open_file_structure('label_structure.txt', True)

# Create lookup table containing multiple information about images
file_lookup_df = pd.DataFrame({'file_name': files})
file_lookup_df = expand_column(file_lookup_df, 'file_name', file_name_components)
```
